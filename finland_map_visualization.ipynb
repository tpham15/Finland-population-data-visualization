```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./data/fi.csv')
cities.head()
```

```python
#Extract Latitude and Longtitude
latd, longd = cities['lat'],cities['lng']
population, area = cities['population'],cities ['population_proper']

plt.style.use('seaborn')
plt.figure(figsize=(14,12))
#Using Pyplot API to plot cities
plt.scatter(longd,latd,
           c = (population), cmap= 'viridis',
           s= (area)/100,linewidths=0, alpha= 0.65)
plt.axis('equal');
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='(population)')

    
    
plt.title('Finland Cities Population Distribution')
```
